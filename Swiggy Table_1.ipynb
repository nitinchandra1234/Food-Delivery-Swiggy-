```python
# Importing all the required libraries for collection and extraction

import requests
import html_to_json
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys


import time
import re
import pandas as pd
```

```python
# Automating the extraction of all restaurants in Bangalore from the Swiggy Website  

home_page = "https://www.swiggy.com/"
driver = webdriver.Chrome()

# Locating the location input bar and storing in location_element
driver.get(home_page)
location_input = driver.find_element(by=By.XPATH, value='//*[@id="location"]')

# Entering the delivery location in the location input bar
location = "Bangalore"
location_input.send_keys(location)
location_input.send_keys(Keys.ENTER)
time.sleep(3)

# clicking the First location after entering Bangalore in the location section 
driver.find_element(by=By.XPATH,
                    value='//*[@id="root"]/div[1]/div[1]/div/div[1]/div[1]/div/div[2]/div/div[2]/button[1]').click()
time.sleep(10)

# Extracting the maximum number of restaurants in the integer format
total_restaurant = driver.find_element(by=By.XPATH,
                                      value='//*[@id="all_restaurants"]/div/div[1]/div/div/div/div[1]').text
total_restaurant = int(total_restaurant.split()[0])

a = 0
b = 500
# Running the while loop to extract the details of all restaurents by keeping the scrooling button at middile. 
while True:
    driver.execute_script(f"window.scrollTo({a}, {b});")
    time.sleep(3)
    ele = driver.find_elements(by=By.XPATH, value='//*[@id="all_restaurants"]/div/div[2]/div[1]/div/div')
    print(len(ele))
    if len(ele) >= (total_restaurant//4):  # Breaking the loop at total_restaurants/4 as each element contains 4 restaurants.
        break
    else:
        a = a + 500
        b = b + 500
    
print("end of while loop")
total_elements = ele
# Once the loop ends we can extract the elements as the scroll has reached the end of the page
elements = driver.find_elements(by=By.XPATH, value='//*[@id="all_restaurants"]/div/div[2]/div[1]/div/div')
```

```python

# Extracting the innerHTML of the elements into a list
html_list = []

# Looping through each elements to get the innerHTML and store it in a list
for i in elements:
    html = i.get_attribute("innerHTML")
    html_list.append(html)

print(html_list)
```

```python
# Verifying the number of restaurant names in a single element of the html_list
ele = html_list[0]

restaurant_name = re.findall('<div class="nA6kb">(.*?)</div>', ele)
restaurant_name
```

```python
# Extracting the required data from the html_list

base_url = "https://www.swiggy.com"
# List to store the restaurant data
restaurant_data = []
k= 0
links = []
# Loop through the html strings and extract the restaurant data
for html in html_list:
#     restaurant_ids = re.findall('data-rest-id="(.*?)"', html)
    restaurant_links = re.findall('<a href="(.*?)" class', html)
    name = re.findall('<div class="nA6kb">(.*?)</div>', html)
    rating = re.findall('<span class="icon-star _537e4"></span><span>(.*?)</span>', html)
    cuisines = re.findall('<div class="_1gURR" title="(.*?)">', html)
    price = re.findall('<div class="nVWSi">(.*?)</div>', html)
    delivery_time = re.findall('<div>(\d+) MINS</div>', html)
    k= k+1
    # If price or cusins not found, set it to empty string
    if not price:
        price = [""]
    if not cuisines:
        cuisines = [""]
    k= k+1
    if k < 367:
    # Add the restaurant data to the list
        restaurant_data.append([(base_url+restaurant_links[0]),name[0], rating[0], cuisines[0], price[0], delivery_time[0]])
        restaurant_data.append([(base_url+restaurant_links[1]),name[1], rating[1], cuisines[1], price[1], delivery_time[1]])
        restaurant_data.append([(base_url+restaurant_links[2]),name[2], rating[2], cuisines[2], price[2], delivery_time[2]])
        restaurant_data.append([(base_url+restaurant_links[3]),name[3], rating[3], cuisines[3], price[3], delivery_time[3]])
     
    else:
        restaurant_data.append([(base_url+restaurant_links[0]),name[0], rating[0], cuisines[0], price[0], delivery_time[0]])
        restaurant_data.append([(base_url+restaurant_links[1]),name[1], rating[1], cuisines[1], price[1], delivery_time[1]])  
      
# Create a dataframe from the restaurant data
df88 = pd.DataFrame(restaurant_data, columns=['Link','Restaurant Name', 'Rating', 'Cuisines', 'Price for Two', 'Delivery Time'])
df88
```
